```python
from rembg import remove 
```

```python
from PIL import Image
```

```python
input_path = "images.jpeg"        # Use the attached file you provided
output_path = "output.png"

input_img = Image.open(input_path)
output_img = remove(input_img)
output_img.save(output_path)    # Correct method to save the output image

```

```python
import numpy as np
```

```python
data = np.array(output_img)

# Create green background: (Red=0, Green=255, Blue=0)
green_bg = np.zeros_like(data)
green_bg[..., 0] = 0    # Red
green_bg[..., 1] = 255  # Green
green_bg[..., 2] = 0    # Blue
green_bg[..., 3] = 255  # Alpha

# Where alpha channel is 0 (original background), use green background
mask = data[..., 3] == 0
data[mask] = green_bg[mask]
```

```python
result_img = Image.fromarray(data)
result_img.save(output_path)
```
